```python
import pandas as pd
import numpy as np

stepData = [3543, 3454, 2348, np.NaN, 3423, 7567, np.NaN, 2348, 5345, 7357, np.NaN, 3242, 9684]
stepData2 = [2359, 5438, 8249, 3458, 6937, 3592, 6543, 5396, 2359, 6924, 6843, 2502, 6483]
stepCounts = pd.Series(stepData, name = 'steps')
stepCounts2 = pd.Series(stepData2, name = 'steps')

stepCounts.index = pd.date_range('20240508', periods=13)

print(stepCounts)
```

```python
# Data Imputation
print(stepCounts.dtypes)
print(stepCounts2, stepCounts2.dtypes)
```

```python
step_float = stepCounts2.astype(np.float64)
#print(step_float, step_float.dtypes)

step_float[1] = np.nan
step_float[4] = np.nan
step_float[7] = np.nan
print(step_float)
```

```python
# Find missing values
for i in step_float[step_float.isna() == False]:# list compression
  print(i)
```

```python
# Replace missing values
step_float.fillna(0)
```
